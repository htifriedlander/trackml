```python
def cherry_picked_data_iter (batch_size):
    for batch in simple_batch_iter (hits, truth, batch_size, ideal = True):
        tracks = np.zeros ([batch_size, 20])
        for i in range(batch_size):
            for j in range(10):
                tracks[i][2*j] = (hits_trans.loc[hits['hit_id'] == batch[i][0][j]])['phi'].values[0]
                tracks[i][2*j+1] = (hits_trans.loc[hits['hit_id'] == batch[i][0][j]])['z'].values[0]
        yield tracks
```
